## 💻 Code-Zelle 1: Bibliotheken und Datenvorbereitung

````python
import numpy as np
import matplotlib.pyplot as plt

# Generiere einen Array mit 100 Werten zwischen 0 und 10
x = np.linspace(0, 10, 100)

# Berechne den Sinus für jeden x-Wert
y = np.sin(x)

print(f"Erfolgreich {len(x)} Datenpunkte generiert!")
````

## 📝 Markdown-Zelle 2: Mathematische Formeln (LaTeX)

## Mathematischer Hintergrund

Jupyter Notebooks unterstützen komplexe mathematische Darstellung. In unserem Code oben verwenden wir eine einfache trigonometrische Funktion: $f(x) = \sin(x)$

Für komplexere Data-Science-Projekte kannst du auch abgesetzte Gleichungen nutzen. Hier ist als Beispiel die Dichtefunktion der Normalverteilung:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Jetzt wollen wir uns ansehen, wie unsere generierten Sinus-Daten visuell aussehen.

## 💻 Code-Zelle 2: Daten visualisieren

````python
# Erstelle den Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sinuskurve', color='royalblue', linewidth=2, linestyle='-')

# Füge Titel und Achsenbeschriftungen hinzu
plt.title('Beispiel-Visualisierung: Sinuswelle', fontsize=14)
plt.xlabel('X-Werte', fontsize=12)
plt.ylabel('Y-Werte (Sinus)', fontsize=12)

# Füge ein Raster und eine Legende hinzu
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Zeige den Graphen an
plt.show()
````
