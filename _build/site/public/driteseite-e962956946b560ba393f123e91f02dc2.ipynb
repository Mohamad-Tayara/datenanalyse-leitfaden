## 2. Die Phasen des Data-Lifecycle-Managements

Ein strukturiertes Datenprojekt lässt sich standardmäßig in fünf Kernphasen unterteilen. Jede Phase baut auf der vorherigen auf, weshalb Fehler in der Anfangsphase oft zu schwerwiegenden Abweichungen im finalen Report führen.

### Phase 1: Datenbeschaffung (Data Ingestion)
Die Datenbeschaffung stellt das Fundament dar. Hierbei werden relationale Datenbanken (z. B. PostgreSQL), NoSQL-Systeme oder flache Dateiformate (`.csv`, `.json`, `.xlsx`) über automatisierte Pipelines angebunden.

* **Herausforderung:** Heterogene Datenstrukturen, inkonsistente Zeitzonen und unterschiedliche Zeichensätze (Encodings wie `UTF-8` vs. `ISO-8859-1`).
* **Best Practice:** Daten direkt beim Import in ein einheitliches Format (z. B. `pandas.DataFrame`) überführen und erste Konsistenzprüfungen durchführen.

### Phase 2: Datenbereinigung und Data Cleaning
Rohdaten sind selten sauber. Sie enthalten fehlende Werte, Duplikate oder fehlerhafte Ausreißer, die durch Systemfehler oder menschliche Fehleingaben entstanden sind.

[Rohdaten-Import]
│
├──> Suche nach Duplikaten ──> [Entfernen]
│
├──> Analyse fehlender Werte (NaN) ──> [Imputation / Löschen]
│
└──> Ausreißer-Erkennung (z. B. IQR-Methode) ──> [Filterung]

### Phase 3: Explorative Datenanalyse (EDA)
In dieser Phase geht es darum, die Struktur der Daten visuell und statistisch zu verstehen. Es werden Korrelationen gesucht, Verteilungen analysiert und erste Hypothesen aufgestellt. Hier kommen Bibliotheken wie `matplotlib`, `seaborn` oder interaktive Frameworks wie `altair` zum Einsatz.

## 3. Mathematische Grundlagen der Datenanalyse

Um fundierte Aussagen über die Qualität und die Trends innerhalb eines Datensatzes treffen zu können, nutzen wir etablierte statistische Kennzahlen.

### 3.1 Lineare Korrelation nach Pearson
Der Korrelationskoeffizient $r$ misst die Stärke und Richtung eines linearen Zusammenhangs zwischen zwei stetigen Variablen. Er liegt immer im Intervall $[-1, 1]$.

$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2 \cdot \sum_{i=1}^{n}(y_i - \bar{y})^2}}$$

* **$r = 1$:** Perfekter positiver linearer Zusammenhang.
* **$r = 0$:** Kein linearer Zusammenhang zwischen den Variablen.
* **$r = -1$:** Perfekter negativer linearer Zusammenhang.

### 3.2 Die Normalverteilung
Viele natürliche und technische Prozesse folgen der Gauß- oder Normalverteilung. Die Wahrscheinlichkeitsdichte wird durch die folgende Funktion beschrieben:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Wobei $\mu$ den Mittelwert (Erwartungswert) und $\sigma$ die Standardabweichung (Streuung) der Daten darstellt. Im Jupyter Book nutzen wir diese Formeln, um dem Leser die theoretischen Hintergründe der anschließenden Machine-Learning-Modelle transparent darzulegen.

## 4.2 Datenaggregation

````python
import numpy as np
import pandas as pd

# 1. Wir erstellen uns selbst künstliche Testdaten (damit der Code läuft)
data = {
    "Datum": ["2026-01-15", "2026-01-16", "2026-02-01", "2026-02-15", "2026-03-01"],
    "Kategorie": ["Elektronik", "Bürobedarf", "Elektronik", "Elektronik", "Bürobedarf"],
    "Umsatz": [549.99, 24.50, 89.00, 120.00, 45.00],
    "Status": ["Erfolgreich", "Storniert", "Erfolgreich", "Erfolgreich", "Erfolgreich"]
}

# Das ist das ursprüngliche DataFrame (df)
df = pd.DataFrame(data)

# 2. Schritt 4.1: Daten filtern (Nur 'Erfolgreich' behalten)
# Hier wird 'df_clean' definiert!
df_clean = df[df["Status"] == "Erfolgreich"].copy()

# Sicherstellen, dass das Datum ein echtes Datetime-Objekt ist
df_clean["Datum"] = pd.to_datetime(df_clean["Datum"])


# 3. Schritt 4.2: Datenaggregation (Umsatz pro Kategorie und Monat)
monats_report = (
    df_clean.groupby(["Kategorie", df_clean["Datum"].dt.to_period("M")])
    .agg(Gesamtumsatz=("Umsatz", "sum"), Anzahl_Verkaeufe=("Umsatz", "count"))
    .reset_index()
)

# 4. Ergebnis im Notebook anzeigen
print("--- Aggregierter Monatsbericht ---")
print(monats_report)
````
