### 🔄 Daten-Pipeline und Workflow

Bevor der Algorithmus trainiert werden kann, durchlaufen die Rohdaten eine mehrstufige Pipeline. Der Prozess ist wie folgt aufgebaut:

1. **Datenbeschaffung (Ingestion)**
   * Import aus der SQL-Datenbank oder lokalen `.csv`-Dateien.
2. **Bereinigung (Preprocessing)**
   * Entfernung von Duplikaten.
   * Imputation fehlender Werte (z. B. Ersetzen von `NaN` durch den Median).
3. **Transformation (Feature Engineering)**
   * Normalisierung numerischer Variablen.
   * One-Hot-Encoding für kategoriale Features.
